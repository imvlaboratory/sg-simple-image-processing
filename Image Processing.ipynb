### 1.   BGR

```python
# import library
import numpy as np
import cv2

image = cv2.imread("opencv.png")

# split channel image
(B, G, R) = cv2.split(image)

cv2.imshow("BGR", image)

# show each channel individually
cv2.imshow("Red", R)
cv2.imshow("Green", G)
cv2.imshow("Blue", B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged = cv2.merge([B, G, R])
cv2.imshow("Merged", merged)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2. RGB

```python
import cv2
img1 = cv2.imread('opencv.png')

# convert BGR ke RGB
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

cv2.imshow('BGR', img1)
cv2.imshow('RGB', img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 3.   Grayscale

```python
import cv2
img1 = cv2.imread('baboon.png')

# convert BGR ke Grayscale
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

cv2.imshow('BGR', img1)
cv2.imshow('grayscale', img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4.   Black & White

```python
img1 = cv2.imread('baboon.png')

# convert BGR ke Black & White
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
thresh, BW = cv2.threshold(img2, 128, 255, cv2.THRESH_BINARY)

cv2.imshow('BGR', img1)
cv2.imshow('Black and White', BW)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 5.   HSV

```python
img = cv2.imread('two-cubes.png')

# convert BGR ke HSV
HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

#(H, S, V) = cv2.split(HSV)
H = HSV [: ,: ,0]
S = HSV [: ,: ,1]
V = HSV [: ,: ,2]

cv2.imshow("BGR", img)

# show each channel individually
cv2.imshow("H", H)
cv2.imshow("S", S)
cv2.imshow("V", V)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged = cv2.merge([H, S, V])
cv2.imshow("HSV", merged)
cv2.imshow("BGR", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 6.   YCbCr

```python
img = cv2.imread('two-cubes.png')

# convert BGR ke YCbCr
YCbCr= cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)

Y = YCbCr[: , : ,0]
Cr = YCbCr[: , : ,1]
Cb = YCbCr[: , : ,2]

cv2.imshow("BGR", img)

# show each channel individually
cv2.imshow("Y", Y)
cv2.imshow("Cb", Cb)
cv2.imshow("Cr", Cr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged = cv2.merge([Y, Cb, Cr])
cv2.imshow("YCrCb", merged)
cv2.imshow("BGR", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 7.   CIELAB

```python
import cv2
img = cv2.imread('two-cubes.png')

# convert BGR ke CIELAB
LAB= cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)

L = LAB[: , : ,0]
A = LAB[: , : ,1]
B = LAB[: , : ,2]

cv2.imshow("BGR", img)

# show each channel individually
cv2.imshow("L", L)
cv2.imshow("A", A)
cv2.imshow("B", B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
merged = cv2.merge([L, A, B])
cv2.imshow("LAB", merged)
cv2.imshow("BGR", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Operasi Morfologi Pada Pengolahan Citra

### 1.   Dilasi (Dilation)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('koin.jpg',0)# 0:grayscale 

# Convert BGR ke Black and White (Biner)
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Structuring Element
kernel = np.ones((5,5),np.uint8)

# Dilasi
dilasi = cv2.dilate(thresh,kernel,iterations =1)
 
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(dilasi,cmap = 'gray')
plt.title('Hasil Dilasi'), plt.xticks([]), plt.yticks([])
plt.show()
```

### 2. Erosi (Erosion)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img = cv2.imread('koin.jpg',0)# 0:grayscale

# Convert BGR ke Black and White (Biner)
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Structuring Element
kernel = np.ones((5,5),np.uint8)

# Erosi
erosi = cv2.erode(thresh,kernel,iterations = 1)
 
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(erosi,cmap = 'gray')
plt.title('Hasil Erosi'), plt.xticks([]), plt.yticks([])
 
plt.show()
```

### 3. Opening (Pembukaan)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img = cv2.imread('koin.jpg',0) # 0:grayscale 

# Convert BGR ke Black and White (Biner)
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Structuring Element
kernel = np.ones((5,5),np.uint8)

# Opening
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN, kernel, iterations=1)
 
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(opening,cmap = 'gray')
plt.title('Hasil Opening'), plt.xticks([]), plt.yticks([])
 
plt.show()
```

### 4. Closing (Penutupan)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img = cv2.imread('koin.jpg',0) # 0:grayscale 

# Convert BGR ke Black and White (Biner)
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Structuring Element
kernel = np.ones((5,5),np.uint8)

# Closing
close = cv2.morphologyEx(thresh,cv2.MORPH_CLOSE, kernel, iterations=1)
 
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(close,cmap = 'gray')
plt.title('Hasil Closing'), plt.xticks([]), plt.yticks([])
 
plt.show()
```

## Image Filtering

### 1. Averaging Filter

```python
import cv2
import numpy as np
  
img = cv2.imread('lena.png')

# Averaging Filter
im1 = cv2.blur(img,(5,5))
im2 = cv2.boxFilter(img, -1, (10, 10), normalize=True)  

cv2.imshow('Image Averaging Filter', np.hstack((im1, im2)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2. Gaussian Filter

```python
import cv2
import numpy as np
  
img = cv2.imread('lena.png')

# Gaussian Filter
dst = cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT) 

cv2.imshow('Image Gaussian Filter', np.hstack((img, dst)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 3. Median Filtering

```python
import cv2
import numpy as np

img = cv2.imread('lena.png')

# Median Filter
dst = cv2.medianBlur(img,7)

cv2.imshow('Image Median Filter', np.hstack((img, dst)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4. Bilateral Filter

```python
import cv2
import numpy as np

img = cv2.imread('lena.png')

# Bilateral Filter
dst = cv2.bilateralFilter(img, 5, 10, 10)

cv2.imshow('Image Bilateral Filter', np.hstack((img, dst)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Edge Detection

```python
import cv2
import numpy as np

img = cv2.imread('lena.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### 1. ROBERT

```python
kernelx = np.array([[1, 0],[0, -1]])
kernely = np.array([[0, 1], [-1, 0]])

img_robertx = cv2.filter2D(gray, -1, kernelx)
img_roberty= cv2.filter2D(gray, -1, kernely)
```

### 2. PREWITT

```python
kernelx = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
kernely = np.array([[1, 1 , 1], [0, 0, 0], [-1, -1, -1]])

img_prewittx = cv2.filter2D(gray, -1, kernelx)
img_prewitty= cv2.filter2D(gray, -1, kernely)
```

### 3. SOBEL

```python
kernelx = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
kernely = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

img_sobelx = cv2.filter2D(gray, -1, kernelx)
img_sobely= cv2.filter2D(gray, -1, kernely)
```

### 4. CANNY

```python
edges = cv2.Canny(gray,100,200)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
plt.subplot(151),plt.imshow(gray,cmap = 'gray')
plt.title('Grayscale'), plt.xticks([]), plt.yticks([])
plt.subplot(152),plt.imshow(img_robertx + img_roberty,cmap = 'gray')
plt.title('Robert'), plt.xticks([]), plt.yticks([])
plt.subplot(153),plt.imshow(img_prewittx + img_prewitty,cmap = 'gray')
plt.title('Prewitt'), plt.xticks([]), plt.yticks([])
plt.subplot(154),plt.imshow(img_sobelx + img_sobely,cmap = 'gray')
plt.title('Sobel'), plt.xticks([]), plt.yticks([])
plt.subplot(155),plt.imshow(edges,cmap = 'gray')
plt.title('Canny'), plt.xticks([]), plt.yticks([])
 
plt.show()
```
